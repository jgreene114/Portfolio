```python
import pandas as pd
```

```python
df = pd.read_csv('sea-surface-temperatures.csv')
```

```python
df
```

```python
for column in ['NINO1+2', 'NINO2', 'NINO3.4', 'NINO4']:
    df[column] = df[column].str.replace('°F', '').astype(float)
```

```python
df['Temperature'] = df[['NINO1+2', 'NINO2', 'NINO3.4', 'NINO4']].mean(axis=1).round(2)
```

```python
df
```

```python
df['Year'] = df['MM/YY'].apply(lambda x: x.split(' ')[1])
df['Month'] = df['MM/YY'].apply(lambda x: x.split(' ')[0])
```

```python
df
```

```python
month_to_num = {'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4, 'May': 5, 'Jun': 6,
                'Jul': 7, 'Aug': 8, 'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12}
```

```python
df['Month'] = df['Month'].apply(lambda x: month_to_num[x])
```

```python
df = df.drop(columns=['MM/YY', 'NINO1+2', 'NINO2', 'NINO3.4', 'NINO4'])
```

```python
df = df[['Year', 'Month', 'Temperature']]
```

```python
df.to_csv('sst.csv', index=False)
```
